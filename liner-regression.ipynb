```python
import numpy as np

# Create a training set with 2 features and 1 output
X_train = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Feature 1
y_train = np.array([3, 3, 7, 6, 11])
```

```python
#plot the training set
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.xlabel('Feature 1')
plt.ylabel('Output')
plt.title('Training Data')
plt.legend()
plt.show()
```

```python
#f(x) = w * x + b
def linear_regression(X, y, learning_rate=0.01, epochs=1000):
    m, n = X.shape
    w = np.zeros(n)  # Initialize weights
    b = 0  # Initialize bias

    for epoch in range(epochs):
        y_pred = np.dot(X, w) + b  # Linear model prediction
        error = y_pred - y  # Calculate error

        # Gradient descent step
        w_gradient = (1/m) * np.dot(X.T, error)
        b_gradient = (1/m) * np.sum(error)

        w -= learning_rate * w_gradient
        b -= learning_rate * b_gradient

    return w, b
```

```python
#prediction
def predict(X, w, b):
    return np.dot(X, w) + b
```

```python
#plot the training set, prediction line, and the predicted values
w, b = linear_regression(X_train, y_train)
y_pred = predict(X_train, w, b)
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_pred, color='red', label='Prediction line')
plt.xlabel('Feature 1')
plt.ylabel('Output')
plt.title('Training Data and Prediction')
plt.legend()
plt.show()
```

```python
#use scikit-learn for comparison
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_sklearn = model.predict(X_train)
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_pred_sklearn, color='green', label='Scikit-learn Prediction line')
plt.xlabel('Feature 1')
plt.ylabel('Output')
plt.title('Training Data and Scikit-learn Prediction')
plt.legend()
plt.show()
```
